# **INTRODUCTION**

## **DATASET BRIEF**

The dataset contains transactions from 2009 to 2012 with a total of 5500 raw data, including order status which is divided into **completed orders**, **returned orders**, and, **canceled orders**.

The dataset is named **dqlab_sales_store** and contains the following columns:
1. OrderID
2. Order Status
3. Customer
4. Order Date
5. Order Quantity
6. Sales
7. Discount %
8. Discount
9. Product Category
10. Product Sub-Category

## **OBJECTIVE**

From the dataset that has been provided, the DQLab store management wants to know

1. Overall performance of DQLab Store from 2009 to 2012 for **number of orders** and **total sales of orders finished**

2. Overall performance of DQLab Store by **product sub-category** which will be compared between 2011 and 2012

3. The effectiveness and efficiency of promotions up to this moment, by measuring the burn rate of promotions carried out as a whole by year

4. The effectiveness and efficiency of promotions up to this moment, by measuring the burn rate of promotions carried out as a whole based on sub-categories (in 2012)

5. Analysis of the number of new customers each year
