# **BACKGROUND**

FinanKu is a fictitious fintech startup that offers its customers savings and loan options. They provide savings, deposits, unsecured loans, credit cards, and car and motorcycle financing.

FinanKu currently has 20,000 customers spread across three major cities in Indonesia: Jakarta, Bandung, and Surabaya. This figure is quite large given that FinanKu has only been in operation for 1.5 years, with 300,000+ customers expected in the next three years.

This rapid development has caused stakeholders in the FinanKu Credit Division to be more cautious in channeling their credit to avoid payment failure, particularly from the Credit Card line, which has an instant approval feature in 1 minute.

# **PROBLEM STATEMENT**

More than 20% of FinanKu's credit card customers failed to pay in the previous year. Because FinanKu's startup scale was still relatively small, business operations were disrupted.

Concerns about FinanKu credit card payment delays, which will harm business. As a result, people who are likely to experience late payments can be predicted more quickly, allowing appropriate strategies for future conditions to be developed.

# **BUSINESS OBJECTIVE**

FinanKu wants to know ahead of time which lending customers are at risk of payment failure.

# **ANALYSIS OBJECTIVE**

Create a model that can accurately predict at least 60% of customers who will have late credit card payments [Accuracy & Recall above 60%].

# **DATASET INFORMATION**

Some information is available from the existing dataset.

---

**1. Customer ID:** Unique Customer ID\
**2. Branch:** Registered Customer Branch Location\
**3. City:** Registered Customer's City Location\
**4. Age:** Customer Age in the Observation Period\
**5. Avg. Annual Income/Month:** Average customer income in one year\
**6. Balance (Q1-Q4):** Customers' settled balance at the end of the quarter\
**7. Num of Products (Q1-Q4):** Number of customer product holdings at the end of the quarter\
**8. HasCrCard (Q1-Q4):** Customer credit card product ownership status at the end of the quarter\
**9. Active Member (Q1-Q4):** Customer active status\
**10. Unpaid Tagging:** Customer status failed to pay/did not fail to pay

# **EXPERIMENT**

**Review Period:**

1. Customers are reviewed for the past year
2. Customers are reviewed for the last 6 months

**Variable Adjustment**:

1. Balance is viewed as an average over a time horizon, with changes visible at the end and beginning of the review.
2. View the average, maximum, and minimum product count holdings during the review period.
3. The status of customer activity is displayed in the form of months.

## 2. Gradient Boosting

```python
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix, classification_report, make_scorer,accuracy_score,precision_score,recall_score,f1_score
```

```python
gbparameter = {'max_depth':[5,10,15],'gamma':[0.0,0.1,0.2,0.3], 'n_estimators':[25,50,75,100],'learning_rate':[0.05,0.1,0.2,0.3], 'scale_pos_weight':[1,3]}
score = {'accuracy':make_scorer(accuracy_score), 'precision':make_scorer(precision_score),'recall':make_scorer(recall_score), 'f1':make_scorer(f1_score)}
```

* **Gamma**: minimum value of loss reduction required when splitting a branch. The bigger the gamma value specified, the more conservative the model constructed will be, and the likelihood of underfitting will increase.
* **Learning_rate**: feature weight adjustment rate. Each iteration of the model generates weights for the characteristics it contains. Learning_rate can be used to help minimize the weight value so that the model being generated does not overfit.
* **Scale_pos_weight**: the weight assigned to the positive (churn) and negative (non-churn) classes. When using an imbalanced dataset, this hyperparameter is useful. The number of dominant classes divided by the number of minority classes is the often-used value.

Experiment 1

```python
GB_Grid = GridSearchCV(XGBClassifier(), gbparameter, cv=5,refit='recall', verbose=0, n_jobs=-1, scoring=score)
```

## 3. Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
parameter = {'max_depth':[5,10,15,20],'max_features':['auto','sqrt','log2'], 'n_estimators':[25,50,75,100,125],'min_samples_split':[2,3,5,7]}
score = {'accuracy':make_scorer(accuracy_score), 'precision':make_scorer(precision_score),'recall':make_scorer(recall_score), 'f1':make_scorer(f1_score)}
```

* **Max_depth**: maximum number of split branches or levels in one tree. The greater the max_depth value set, the more precise the model will be in classifying data into a class. However, the larger the max_depth value will also make the model overfitting.
* **Max_features**: maximum number of features to take into account while splitting nodes. Similarly with Max_depth, the bigger the number of features examined in branch splitting, the more comprehensive the findings obtained, but the model will overfit to the training data.
* **N_estimator**: the number of trees to construct. Given that Random Forest uses the majority vote principle in categorization, the more trees that are formed, the higher the level of accuracy gained. The more trees that are constructed, however, the more computation time is required.
* **Min_sample_split**: the number of data samples that must be split on an internal node. A large value can cause the model to be more conservatively developed. However, if it is too large, the model being made will be too small.

# **CONCLUSION**

All of the models have an average accuracy of more than 60% but a recall of less than 40% (*Except for the Gradient Boosting Model in Experiment*). It implies that there are still a large number of customers who have the ability to default but are not expected to do so. As a result, the intended goals have not yet been met in this model development cycle.

Future development solutions that can be implemented include:
1. Increase the sample (number of customers) assuming the currently available dataset is not the total customer population
2. Oversampling the minority class (defaulters) so that model building is not biased
3. Expanding time horizons
4. Try variations of other variables (adding new variables, or removing variables that have a low importance value in the last result)
5. Trying to expand hyperparameter combinations in model building
6. Try other supervised machine-learning algorithms.
